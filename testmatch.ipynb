## Loading The Dataset

```python
df=pd.read_csv('test.csv')
df
```

## Missing Values

```python
(df.isnull().sum()/len(df))*100
```

* There are so many missing value present in  column Unnamed:11 . so we have to drop this column. Also we do not required the column named Unnamed:0, So we will also drop this column as well.

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.drop('Unnamed: 11',axis=1,inplace=True)
df
```

```python
df.rename(columns={'100':'Centuries','50':'Half centuries','0':'Zeros'},inplace=True)
df
```

## Remove ( from column player and split into region column)

```python
df[['Player', 'Region']] = df['Player'].str.split('(', n=1, expand=True)
df.head(5)
```

```python
df['Region']=df['Region'].str.replace(')','')
df
```

## Finding Null values

```python
df[df['Runs'].isna()==1]
```

```python
df[df['Half centuries'].isna()==1]
```

```python
df[df['Zeros'].isna()==1]
```

* We can see that null values are present in Runs,half centuries and Zeros column.

## Filling null values with 0

```python
df['Runs']=df['Runs'].fillna(0)
df['Half centuries']=df['Half centuries'].fillna(0)
df['Zeros']=df['Zeros'].fillna(0)
df
```

* We fill the null values with 0

## Spliting the Span column

```python
df[['Start', 'End']] = df['Span'].str.split("-", n=1, expand=True)
df.head(5)
```

* We split the Span column into Start and End.

```python
df['Runs']=df['Runs'].str.replace('-','0')
df['HS']=df['HS'].str.replace('*','')
df['Ave']=df['Ave'].str.replace('-','0')
df['Inns']=df['Inns'].str.replace('-','0')
df['NO']=df['NO'].str.replace('-','0')
df['HS']=df['HS'].str.replace('-','0')
df['Centuries']=df['Centuries'].str.replace('-','0')
df['Half centuries']=df['Half centuries'].str.replace('-','0')
df['Zeros']=df['Zeros'].str.replace('-','0')
df
```
